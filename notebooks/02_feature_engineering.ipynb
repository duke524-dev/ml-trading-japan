```python
# åŠ è½½æ•°æ®
data_path = Path.cwd().parent / 'data' / 'TOPIX_18months_clean.csv'
if data_path.exists():
    df = pd.read_csv(data_path, nrows=10000)
    df['datetime'] = pd.to_datetime(df['datetime'])
    print(f'âœ… åŠ è½½ {len(df):,} è¡ŒçœŸå®æ•°æ®')
else:
    print('âš ï¸ æ•°æ®æ–‡ä»¶ä¸å­˜åœ¨ï¼Œä½¿ç”¨ç¤ºä¾‹æ•°æ®')
    dates = pd.date_range('2024-01-01', periods=1000, freq='1min')
    df = pd.DataFrame({
        'datetime': dates,
        'open': 2400 + np.cumsum(np.random.randn(1000) * 2),
        'high': 2400 + np.cumsum(np.random.randn(1000) * 2) + 5,
        'low': 2400 + np.cumsum(np.random.randn(1000) * 2) - 5,
        'close': 2400 + np.cumsum(np.random.randn(1000) * 2),
        'volume': np.random.randint(1000, 50000, 1000)
    })
```

```python
# åˆ›å»ºç‰¹å¾
engineer = FeatureEngineer()
df_features = engineer.create_all_features(df)
df_features = df_features.dropna()

print(f'ç‰¹å¾æ•°é‡: {len(df_features.columns)}')
print(f'æ ·æœ¬æ•°é‡: {len(df_features):,}')
```
